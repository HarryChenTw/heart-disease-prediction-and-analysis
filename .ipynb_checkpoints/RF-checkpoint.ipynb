```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split 
from sklearn.ensemble import RandomForestClassifier 

#for model evaluation
from sklearn.metrics import confusion_matrix, roc_curve, auc 

#for plot
import matplotlib.pyplot as plt
import seaborn as sns #for plotting

#for explanation
import eli5 
from eli5.sklearn import PermutationImportance
import shap #for SHAP values
from pdpbox import pdp, info_plots #for partial plots

pd.options.mode.chained_assignment = None  #hide any pandas warnings
```

```python
dt = pd.read_csv("data/heart.csv")
dt.head(5)
```

## Data Processing

```python
# give each type a real name
dt.columns = ['age', 'sex_male', 'chest_pain_type', 'resting_blood_pressure', 'cholesterol', 'fasting_blood_sugar_greater than 120mg/ml', 'rest_ecg', 'max_heart_rate_achieved',
       'exercise_induced_angina', 'st_depression', 'st_slope', 'num_major_vessels', 'thalassemia', 'target']

dt['chest_pain_type'][dt['chest_pain_type'] == 0] = 'typical_angina'
dt['chest_pain_type'][dt['chest_pain_type'] == 1] = 'atypical_angina'
dt['chest_pain_type'][dt['chest_pain_type'] == 2] = 'non-anginal_pain'
dt['chest_pain_type'][dt['chest_pain_type'] == 3] = 'asymptomatic'

dt['rest_ecg'][dt['rest_ecg'] == 0] = 'normal'
dt['rest_ecg'][dt['rest_ecg'] == 1] = 'ST-T_wave_abnormality'
dt['rest_ecg'][dt['rest_ecg'] == 2] = 'left_ventricular_hypertrophy'

dt['st_slope'][dt['st_slope'] == 0] = 'upsloping'
dt['st_slope'][dt['st_slope'] == 1] = 'flat'
dt['st_slope'][dt['st_slope'] == 2] = 'downsloping'

dt['thalassemia'][dt['thalassemia'] == 1] = 'normal'
dt['thalassemia'][dt['thalassemia'] == 2] = 'fixed_defect'
dt['thalassemia'][dt['thalassemia'] == 3] = 'reversable_defect'

dt['chest_pain_type'] = dt['chest_pain_type'].astype('object')
dt['rest_ecg'] = dt['rest_ecg'].astype('object')
dt['st_slope'] = dt['st_slope'].astype('object')
dt['thalassemia'] = dt['thalassemia'].astype('object')

# Creating Dummy Variables (one-hot encode)
dt = pd.get_dummies(dt)
dt.head(5)
```

```python
y = dt.target
x_data = dt.drop(['target'], axis = 1)

# Normalize
#x = (x_data - np.min(x_data)) / (np.max(x_data) - np.min(x_data)).values

#split the data
seed = 333
X_train, X_test, y_train, y_test = train_test_split(x_data, y, test_size = .2, random_state = seed) 
print('Training size : ', X_train.shape[0])
print('Testing size : ', X_test.shape[0])
```

## Model

```python
model = RandomForestClassifier(n_estimators = 100, random_state = seed, max_depth = 5)
model.fit(X_train, y_train)
```

## Model Evaluation

```python
y_predict = model.predict(X_test)
y_pred_quant = model.predict_proba(X_test)[:, 1]
y_pred_bin = model.predict(X_test)

acc = model.score(X_test,y_test)*100
print("Accuracy Score : {:.2f}%\n\n".format(acc))

CM = confusion_matrix(y_test, y_pred_bin)
print('confusion_matrix : \n', CM)

sensitivity = CM[0,0]/(CM[0,0]+CM[1,0])
print('Sensitivity : ', sensitivity )

specificity = CM[1,1]/(CM[1,1]+CM[0,1])
print('Specificity : ', specificity)

print('auc(fpr, tpr) : ', auc(fpr, tpr))

#plot roc curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_quant)
fig, ax = plt.subplots()
ax.plot(fpr, tpr)
ax.plot([0, 1], [0, 1], transform=ax.transAxes, ls="--", c=".3")
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.rcParams['font.size'] = 12
plt.title('ROC curve for diabetes classifier')
plt.xlabel('False Positive Rate (1 - Specificity)')
plt.ylabel('True Positive Rate (Sensitivity)')
plt.grid(True)
```

```python
df = pd.read_csv("data/heart.csv")

a = pd.get_dummies(df['cp'], prefix = "cp")
b = pd.get_dummies(df['thal'], prefix = "thal")
c = pd.get_dummies(df['slope'], prefix = "slope")
frames = [df, a, b, c]
df = pd.concat(frames, axis = 1)
df = df.drop(columns = ['cp', 'thal', 'slope'])


y = df.target
x_data = df.drop(['target'], axis = 1)

# Normalize
x = (x_data - np.min(x_data)) / (np.max(x_data) - np.min(x_data)).values

X_train, X_test, y_train, y_test = train_test_split(x_data, y, test_size = .2, random_state = 0) 
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 40)  # n_neighbors means k
knn.fit(X_train, y_train)
prediction = knn.predict(X_test)

print("{} NN Score: {:.2f}%".format(40, knn.score(X_test, y_test)*100))
```

```python
from sklearn.svm import LinearSVC

svm = LinearSVC()
svm.fit(X_train, y_train)

acc = svm.score(X_test,y_test)*100
print("Test Accuracy of SVM Algorithm: {:.2f}%".format(acc))
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_train, y_train)

acc = nb.score(X_test,y_test)*100
print("Accuracy of Naive Bayes: {:.2f}%".format(acc))
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train,y_train)
acc = lr.score(X_test,y_test)*100

print("Test Accuracy {:.2f}%".format(acc))
```
